#### Finding optimum K value using Elbow method

```python
x = df.iloc[:, [0, 1, 2, 3]].values


from sklearn.cluster import KMeans

wcss=[]
for i in range(1,11):
    model = KMeans(n_clusters=i,init = 'k-means++',max_iter = 400, n_init = 10, random_state = 0)
    model.fit(x)
    wcss.append(model.inertia_)
    

# Elbow method to find out optimum k value
plt.figure(figsize=(10,6))
plt.plot(range(1,11), wcss)
plt.title('Elbow method(for Optimum K)')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

#### K means clustering , predicting clusters for data points and visualizing them

```python
model = KMeans(n_clusters = 3)
model_pred = model.fit_predict(x)
model.fit(x)
print(model.labels_)
```

```python
print(model.labels_)
plt.figure(figsize=(8, 6))
plt.scatter(x[model_pred  == 0, 2], x[model_pred == 0, 3], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[model_pred  == 1, 2], x[model_pred == 1, 3], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[model_pred  == 2, 2], x[model_pred  == 2, 3],s = 100, c = 'Green', label = 'Iris-virginica')
plt.scatter(model.cluster_centers_[:, 2], model.cluster_centers_[:,3], s = 100, c = 'yellow', label = 'Centroids')
plt.xlabel("Petal Length in cm")
plt.ylabel("Petal Width in cm")
plt.title("KMeans Clustering of Iris Flower types based on Petal Length and Width")
plt.legend()
plt.show()
```

#### Predicting cluster for test input and evaluating the model

```python
print("Actual:",iris.target,"")
print("Predicted:",model.labels_)
```

```python
accuracy = accuracy_score(iris.target,model.labels_)

print(accuracy)
```
